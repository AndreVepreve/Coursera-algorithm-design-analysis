```
# Global settings
import requests
import bisect
```

```
def loadData(url):
  data = []
  response = requests.get(url)
  for line in (line for line in response.iter_lines() if line != ''):
    data.append(int(line))
  return data
```

```
def twoSum(array):
  WIDTH = 10000
  out = set()
  for i in array:
    lower = bisect.bisect_left(array, -WIDTH - i)
    upper = bisect.bisect_right(array, WIDTH - i)
    out |= set([i + j for j in array[lower:upper]])
  return out
```

```
#Download and test assignment data

url = "https://raw.githubusercontent.com/AndreVepreve/Coursera/master/Algorithms-Design-and-Analysis-Part-1-and-2/Data/algo1-programming_prob-2sum.txt"
array = loadData(url)
array.sort()

assert len(array) == 1000000
assert array[0] == -99999887310
assert array[-1] == 99999662302

print 'Original file has %d entries' % len(array)
```

```
# Compute 2SUM count
numberOfValues = len(twoSum(array))

assert numberOfValues == 427

print 'Number of target values found: %d' % numberOfValues
```

```
def maintainMedian(stream):
  out = [stream.pop(0)]
  heapLo, heapHi = [out[0]], []
  while stream:
    num = stream.pop(0)
    if num >= heapLo[0]:
      heapInsert(heapHi, num, 'min')
    else:
      heapInsert(heapLo, num, 'max')
    if len(heapLo) > len(heapHi) + 1:
      heapInsert(heapHi, heapExtract(heapLo, 'max'), 'min')
    if len(heapHi) > len(heapLo) + 1:
      heapInsert(heapLo, heapExtract(heapHi, 'min'), 'max')
    if len(heapLo) >= len(heapHi):
      out.append(heapLo[0])
    else:
      out.append(heapHi[0])
  return out
```

```
# Heap support

def bubbleUp(array, index, kind):
  parentIndex = (index - 1) // 2
  option = {'min': '>', 'max': '<'}
  minOrMax = 'array[index] %s= array[parentIndex]' % option[kind]
  if index == 0 or eval(minOrMax):
    return
  else:
    array[parentIndex], array[index] = array[index], array[parentIndex]
    bubbleUp(array, parentIndex, kind)

def bubbleDown(array, index, kind):
  option = {'min': '<', 'max': '>'}
  minOrMax = 'array[index] %s= array[childIndex]' % option[kind]
  try:
    childIndex = array.index(eval(
      '%s(array[index * 2 + 1], array[index * 2 + 2])' % kind))
  except:
    if index * 2 + 2 == len(array):
      childIndex = len(array) - 1
    else:
      return
  if eval(minOrMax):
    return
  else:
    array[childIndex], array[index] = array[index], array[childIndex]
    bubbleDown(array, childIndex, kind)

def heapInsert(heap, item, kind):
  heap.append(item)
  bubbleUp(heap, len(heap) - 1, kind)

def heapExtract(heap, kind):
  heap[0], heap[-1] = heap[-1], heap[0]
  out = heap.pop()
  if len(heap):
    bubbleDown(heap, 0, kind)
  return out
```

```
#Download and test assignment data

url = "https://raw.githubusercontent.com/AndreVepreve/Coursera/master/Algorithms-Design-and-Analysis-Part-1-and-2/Data/Median.txt"
stream = loadData(url)

assert len(stream) == 10000
assert stream[0] == 6331
assert stream[-1] == 5940

print 'Original file has %d entries' % len(stream)
```

```
# Compute sum of medians

medians = maintainMedian(stream)
sumOfMedians = sum(medians) % 10000

assert sumOfMedians == 1213

print 'Sum of medians is %d' % sumOfMedians
```
