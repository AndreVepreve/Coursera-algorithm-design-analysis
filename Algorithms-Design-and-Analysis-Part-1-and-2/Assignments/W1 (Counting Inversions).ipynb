```
def sortAndCountInversions(intArray, start, end):
  if (start == end):
    return 0
  else:
    splitPoint = (end + start)//2
    x = sortAndCountInversions(intArray, start, splitPoint)
    y = sortAndCountInversions(intArray, splitPoint + 1, end)
    z = mergeAndCountSplitInversions(intArray, start, end, splitPoint + 1)
    return (x+y+z)
```

```
def countInversions(intArray):
  return sortAndCountInversions(intArray, 0, len(intArray) - 1)
```

```
def mergeAndCountSplitInversions(intArray, start, end, splitPoint):
  i = start
  j = splitPoint
  sortedArray = []
  invCount = 0

  while (i < splitPoint or j <= end):
      if (i == splitPoint):
          sortedArray.append(intArray[j])
          j += 1
      elif (j > end):
          sortedArray.append(intArray[i])
          i += 1
      elif (intArray[i] <= intArray[j]):
          sortedArray.append(intArray[i])
          i += 1
      else:
          invCount += (splitPoint - i) 
          sortedArray.append(intArray[j])
          j += 1

  i = start
  while (i <= end):
      intArray[i] = sortedArray[i - start]
      i += 1

  return invCount
```

```
#Test Cases
assert countInversions([1,2,3,4,5,6]) == 0
assert countInversions([6,5,4,3,2,1]) == 15
assert countInversions([1,3,5,2,4,6]) == 3
```
