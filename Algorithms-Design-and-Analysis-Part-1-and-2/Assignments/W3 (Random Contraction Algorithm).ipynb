```
import collections
import copy
import math
import random
```

```
def edgeContraction(dic, nodeA, nodeB):
  dic[nodeA] = dic[nodeA] + dic[nodeB]
  dic.pop(nodeB)
  for i in dic:
    for j in range(len(dic[i])):
      if dic[i][j] == nodeB:
        dic[i][j] = nodeA
  dic[nodeA] = list(filter(lambda x: x != nodeA, dic[nodeA]))
  return dic
```

```
def computeMinCut(dic):
  if len(dic) == 2:
    return len(list(dic.values())[0])
  else:
    nodeA = random.choice(list(dic.keys()))
    nodeB = random.choice(dic[nodeA])
    computeMinCut(edgeContraction(dic, nodeA, nodeB))
    return len(list(dic.values())[0])
```

```
#Download and test assignment data

import requests
url = "https://raw.githubusercontent.com/AndreVepreve/Coursera/master/Algorithms-Design-and-Analysis-Part-1-and-2/Data/kargerMinCut.txt"
response = requests.get(url)

original_graph = {}
for line in (line for line in response.iter_lines() if line != ''):
  dic = list(map(int,line.strip().split("\t")))
  original_graph[dic.pop(0)] = dic
    
print 'Original graph has %d elements' % len(original_graph)

assert original_graph[1][0] == 37
assert original_graph[1][-1] == 196
```

```
i = 0
length = len(original_graph)
mincuts = length**2
while i < length:
  graph = copy.deepcopy(original_graph)
  newMinCut = computeMinCut(graph)
  if newMinCut < mincuts:
    mincuts = newMinCut
    print(mincuts)
  i += 1
```
