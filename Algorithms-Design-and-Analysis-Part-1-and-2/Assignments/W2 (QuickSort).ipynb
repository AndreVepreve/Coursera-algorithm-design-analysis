### ** Programming Assignment #2 **
Master Method & QuickSort algorithm

```
import sys
sys.setrecursionlimit(10000)

PIVOT_FIRST = 1
PIVOT_LAST = 2
PIVOT_MEDIAN = 3

globalComparisons = 0
```

```
def swap(array, i, j):
  temp = array[i]
  array[i] = array[j]
  array[j] = temp
```

```
def isMedian(array, i, j, k):
  return ((array[i] < array[j] and array[i] > array[k]) or
          (array[i] > array[j] and array[i] < array[k]))
```

```
def quickSort(array, left, right, pivotType):
  if left >= right:
    return

  pivot = 0
  global globalComparisons
  if pivotType == PIVOT_FIRST:
    pivot = array[left]
  elif pivotType == PIVOT_LAST:
    pivot = array[right]
    swap(array, left, right)
  elif pivotType == PIVOT_MEDIAN:
    median = left + ((right-left) >> 1)
    if isMedian(array, left, median, right):
      pivot = array[left]
    elif isMedian(array, median, left, right):
      pivot = array[median]
      swap(array, left, median)
    else:
      pivot = array[right]
      swap(array, left, right)
  globalComparisons += (right-left)

  i = left+1
  for j in range(left + 1, right + 1):
    if array[j] < pivot:
      swap(array, i, j)
      i += 1
  swap(array, left, i - 1)

  quickSort(array, left, i - 2, pivotType)
  quickSort(array, i, right, pivotType)
```

```
def runQuickSort(array, pivotType):
  quickSort(array, 0, len(array) - 1, pivotType)
```

```
#Test Cases

inputArray = [1,3,5,2,4,6]
runQuickSort(inputArray, PIVOT_FIRST)
assert inputArray == [1, 2, 3, 4, 5, 6]

inputArray = [1,3,5,2,4,6]
runQuickSort(inputArray, PIVOT_LAST)
assert inputArray == [1, 2, 3, 4, 5, 6]

inputArray = [1,3,5,2,4,6]
runQuickSort(inputArray, PIVOT_MEDIAN)
assert inputArray == [1, 2, 3, 4, 5, 6]
```
